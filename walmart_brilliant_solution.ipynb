```python
all_shipment_data=pd.DataFrame()
origin_warehouse=df2['origin_warehouse']
all_origin_warehouse=[]
#count=0
shipment_identifier=df1['shipment_identifier']
shipments=df2['shipment_identifier']
for index,value in enumerate(shipments):
    for i,data in enumerate(shipment_identifier):
        if value==data:
           # print(origin_warehouse[index])
            #count=count+1
            all_origin_warehouse.append(origin_warehouse[index])
#print(count)
all_origin_warehouse
```

```python
destination_store=df2['destination_store']
all_destination_store=[]
shipment_identifier=df1['shipment_identifier']
shipments=df2['shipment_identifier']
for index,value in enumerate(shipments):
    for i,data in enumerate(shipment_identifier):
        if value==data:
            all_destination_store.append(destination_store[index]) 
all_destination_store 
```

```python
driver_identifier=df2['driver_identifier']
all_driver_identifier=[]
shipment_identifier=df1['shipment_identifier']
shipments=df2['shipment_identifier']
for index,value in enumerate(shipments):
    for i,data in enumerate(shipment_identifier):
        if value==data:
            all_driver_identifier.append(driver_identifier[index])
all_driver_identifier
```

```python
all_shipment_data['origin_warehouse']=all_origin_warehouse
all_shipment_data['destination_store']=all_destination_store
all_shipment_data['driver_identifier']=all_driver_identifier
all_shipment_data['product']=df1['product']
all_shipment_data['on_time']=df1['on_time']
```

```python
all_shipment_data
```

```python
p=df1['product'].tolist()
```

```python
df1[['True','False']]=pd.get_dummies(df1['on_time'])
```

```python
q=0
quantity=[]
T=df1['True']
for j,value in enumerate(p):
    print(value)
    if(j<109):
        if((p[j]==p[j+1])&(T[j]==T[j+1])):
            q=q+1
            
    if(j<109):        
        if(p[j]!=p[j+1]):
            print(q)
            quantity.append(q+1) 
            q=0
    if(j==109):
        quantity.append(q+1)
quantity   
```

```python
all_shipment_data.drop_duplicates(inplace=True) 
```

```python
all_shipment_data['product_quantity']=quantity
```

```python
all_shipment_data.head()
```

```python
frames=[df0,all_shipment_data]
all_data=pd.concat(frames,ignore_index=True)
```

```python
qa=all_data['product_quantity'].tolist()
all_data[['True','False']]=pd.get_dummies(all_data['on_time'])
```

```python
T=all_data['True']
m=all_data['product'].tolist()
o=all_data['origin_warehouse'].tolist()
new=[]
for j,value in enumerate(qa):
    print(value)
    if(j<153):
        if((m[j]==m[j+1])&(T[j]==T[j+1])&(o[j]==o[j+1])):
            new.append(qa[j]+qa[j+1])
            #print(1)
    if(j<153):        
        if((m[j]!=m[j+1])):
            new.append(qa[j])
            #print(1)
    
    if(j==153):
        new.append(qa[j])
new  
```
